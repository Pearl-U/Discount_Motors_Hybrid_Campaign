## Exploratory Analysis

```python
plt.hist(x='price', data=df_toyota)
plt.show()
```

From the histogram plotted below of the price distribution of all cars sold in the past 6 months, it can be seen that the distribution is skewed to the right which means that most vehicles sell for between 8000 - 10000 GBP. With an upper limit of around 30000 GBP, anything above that can be considered outliers

```python
sns.set_style('whitegrid')
sns.histplot(x='price', data=df_toyota)
plt.ylabel('Number of vehicles sold')
plt.title("Distribution of number of cars sold at a given price ")
plt.show()
```

### How the sales differ between Hybrid cars and other fuel types?
From the plot below, it is clear that petrol types are the most common type of vehicles sold. Hybrid vehicle sales are about half the sales of petrol car sales. This leaves a lot of room for improvement of hybrid car sales.

```python
sort=['Petrol','Hybrid','Diesel','Other']
sns.set_style('whitegrid')
sns.countplot(x='fuelType',data=df_toyota, order=sort)
plt.ylabel('Number of vehicles sold')
plt.title("Number of cars sold based on fuel type ")
plt.show()
```

```python
sns.scatterplot(x='year',y='price', hue='fuelType', alpha=0.4, data=df_toyota)
```

```python
hybrid_df = df_toyota[df_toyota['fuelType'] == 'Hybrid']
hybrid_df.head()
```

```python
sns.scatterplot(x='year',y='price', hue='fuelType', alpha=0.4, data=hybrid_df)
plt.ylabel('price of vehicles sold')
plt.title("Number of cars sold based on registration year ")
plt.show()
```

Vehicle puchases appear to be mostly based on price and from the graph above, hybrid vehicles increase in price as the registration year increases. To increases competitiveness among other fuel types, hybrid vehicles under 15000 across all years should be considered for the campaign.

### How to make Hybrid cars more attractive?

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='tax',hue='fuelType',data=df_toyota)
plt.ylabel('Vehicles sold')
plt.title("Number of cars sold by fuel type and road tax ")
plt.xticks(rotation=90)
plt.show()
```

Vehicles with 0 road tax are mostly Hybrid vehicles. This can be used as a selling point to attract more customers.

```python

sns.barplot(x='model', y='tax', data=hybrid_df, errorbar=None)
plt.ylabel('Road tax (GBP)')
plt.title("Road tax on hybrid models ")
plt.show()
```

The Auris and Yaris are the hybrid cars with the lowest tax. These can be the main vehicles used in the campaign.

```python
sns.barplot(x='model', y='price', data=hybrid_df, errorbar=None)
plt.ylabel('Price (GBP)')
plt.title("Prices of hybrid models sold ")
plt.show()
```

Yaris and Auris are also the cheapest Hybrid vehicles.

```python
sns.countplot(x='model', data=hybrid_df)
plt.ylabel('Number of vehicles sold')
plt.title("hybrid models sold ")
plt.show()
```

Based on sales Yaris, Auris and C-HR are the top 3 Hybrid models sold. Prius could be advertised more, along side Yaris and Auris, as it is cheaper than C-HR and its road tax is also less than C-HR.

```python
sns.boxplot(x='model', y='price', data=hybrid_df)
plt.ylabel('Price of vehicles sold (GBP)')
plt.title("Distribution of hybrid models sold ")
plt.show()
```

Yaris, Prius and Auris are the only models that are priced under 10000 GBP. These hybrid models can rival the petrol models in this price range.

```python
sns.boxplot(x='fuelType', y='price', data=df_toyota)
plt.ylabel('Price of vehicles sold (GBP)')
plt.title("Distribution of cars sold by fuel type ")
#plt.ylim(0,40000)
plt.show()
```

Petrol vehicle sales ranges between 9000 to 12000 GBP and Hybrid vehicle sales ranged between 13000 to 21000 GBP.
